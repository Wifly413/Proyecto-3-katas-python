1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados

```python
def contar_frecuencias(texto):
    """
    Función que se encarga de contar la frecuencia de las letras de un texto en concreto.
    Los espacios no deben ser considerados.

    Args:
      texto: Cadena de letras (str)

    return: dict con frecuencias por letra
    """
    frecuencias = {}
    for letra in texto :
        if letra in frecuencias:
            frecuencias[letra] += 1

        else:
            frecuencias[letra] = 1
    return frecuencias

texto = "Telescopio"
contar_frecuencias(texto)        
```

```python
def contiene_palabras(palabras,p_objetivo):
    """
    Recibe una lista de palabras y una palabras objetivo

    Args:
        palabras : lista de diferentes palabras (str)
        p_objetivo : palabra clave que deben contener la lista de palabras (str)

    returns :

    Lista_nueva : lista que contenga todas las palabras de la original que contengan la palabra objetivo
    """
    lista_nueva = []

    for palabra_individual in palabras:
        
        if p_objetivo in palabra_individual :
           
           lista_nueva.append(palabra_individual)
    
    return lista_nueva

lista_palabras = ["dermatologia", "oncologia", "playa", "asteroide", "traumatologia", "piragua"]
objetivo = "gia"

contiene_palabras(lista_palabras,objetivo)
```

5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5). La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
def notas(nota, nota_aprovado=5):
    """
    Calcula la media de una lista de números y devuelve una tupla
    (media, estado) donde estado es "aprobado" si la media >= nota_aprovado,
    en caso contrario "suspenso".

    Args:
        nota: lista de números (list)
        nota_aprovado: umbral de aprobado (int o float), por defecto 5

    Returns:
        (media, estado)
    """
    media = sum(nota) / len(nota)
    
    if media >= nota_aprovado:
        estado = "aprobado"
    else:
        estado = "suspenso"

    return media, estado

nota = [5.6, 4.7, 8.9, 9.9, 2.4, 7.3]
resultado = notas(nota)
print(resultado)
```

```python
def factorial(n):
    """
    Calcula el factorial de un numero de manera recursiva

    Args:
        n (int): numero del cual se parte
    
    Returns:
        int: el factorial de n
    """
    # De tanto tiempo ni me acordaba que era el factorial y la verdad que es mas sencillo de lo que parecia
    if n <= 1:
        return 1
   
    else:
        return n * factorial(n - 1)

resultado = factorial(5)
print(resultado)
```

13. Genera una función que, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas. Usa la función map().

```python
def mayus_minus(caracteres): 
    """
    Función que recibe una lista de caracteres y devuelve una lista de tuplas, donde cada tupla contiene el carácter en mayúscula y minúscula.
    Args:
        caracteres: lista de caracteres (list of str)
    Returns:
        list: lista de tuplas con el carácter en mayúscula y minúscula
    """

    no_repetidos = []
    for caracter in caracteres :
        if caracter not in no_repetidos:
            no_repetidos.append(caracter)
    
    def cambio(letra):
        return (letra.upper(), letra.lower())
    
    lista_mayus_minus = list(map(cambio, no_repetidos))
    return lista_mayus_minus
    

conjunto = "La vida es corta"
print(mayus_minus(conjunto))
```

28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

```python

def duplicado(dupli):
    """
    Función que recibe una lista y devuelve el primer elemento duplicado que encuentra. 
    Si no hay duplicados, devuelve None.
    Args:
        dupli: lista de elementos (list)
    Returns:
        El primer elemento duplicado encontrado o None si no hay duplicados
    """
    n = len(dupli)
    for indice in range(n):
        for j in range(0, indice):
            if dupli[indice] == dupli[j]:
                return dupli[indice]
            
    return None 

duplicado_lista = ["Oso", "Tigre", "Nutria", "Tigre", "Avestruz", "Oso"]
print(duplicado(duplicado_lista))
```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

```python
def es_anagrama(palabra1, palabra2):
    """
    Función que verifica si dos palabras son anagramas, es decir, si contienen las mismas letras en diferente orden, sin importar mayúsculas o minúsculas.
    Args:
        palabra1: primera palabra (str)
        palabra2: segunda palabra (str)
     Returns:
        bool: True si son anagramas, False si no lo son
     """
    
    p1 = palabra1.lower()
    p2 = palabra2.lower()
    
    return sorted(p1) == sorted(p2)


print(es_anagrama("Amor", "Roma"))    
print(es_anagrama("Hola", "Lola"))    
```

39. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo", "circulo" o "triangulo") y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math

def calculo_area(figura,datos):
    
    if figura == "triangulo" :
        base , altura = datos    
        return base * altura / 2
    elif figura == "circulo":
        radio = datos[0]
        return math.pi * (radio**2)
    elif figura == "rectangulo":
        base , altura = datos
        return base * altura



datos_triangulo= (14,5)
datos_rectangulo= (10,3)
datos_circulo= (10,)       

print("La base del triangulo es" ,calculo_area("triangulo",datos_triangulo))
print("------------------------------")
print("La base del rectangulo es",calculo_area("rectangulo",datos_rectangulo))
print("------------------------------")
print("La base del circulo es",calculo_area("circulo",datos_circulo))
```
